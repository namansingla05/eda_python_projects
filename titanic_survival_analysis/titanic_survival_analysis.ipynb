```python
#now their in two columns few value are missing so we can fill that values
#Handle missing values
data["Embarked"].mode()
```

```python
#catogorical value encoding
#because in the future we need to do mathematical task on that columns so thats why we need this

data["Gender"]=data["Sex"].map({"male":1,"female":0})
data.head()
```
